```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.arange(0, 2, 0.05)
signal_1 = 2*np.sin(2*np.pi*5*x)
signal_2 = np.sin(2*np.pi*12.5*x)
signal_3 = 1.5*np.sin(2*np.pi*20*x)
signal_4 = 0.5*np.sin(2*np.pi*35*x)
```

```python
y= .25+signal_1 + signal_2 + signal_3 + signal_4
```

```python
plt.plot(x, y)
plt.show()
```
